```python
import torch
import torch.nn as nn

class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.seq_modules = nn.Sequential(
            nn.Linear(17, 10),
            nn.ReLU(),
            nn.Linear(10, 5),
            nn.ReLU(),
            nn.Linear(5, 1)
        )

    def forward(self, x):
        x = self.flatten(x)
        y = self.seq_modules(x)
        return y
    
myFirstModel = MyModel()
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(myFirstModel.parameters(), lr=0.01)

for epoch in range(100):
    inputs = torch.randn(1, 17)
    targets = torch.randn(1, 1)
    
    optimizer.zero_grad()
    outputs = myFirstModel(inputs)
    loss = loss_fn(outputs, targets)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
    
```
