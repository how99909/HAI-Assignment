```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 0.5, 1, 2, 3, 4, 4.5, 5, 5.5, 6, 7, 8, 9, 9.5, 10])
y = np.array([60, 62, 63, 68, 76, 83, 87, 89, 92, 94, 97, 99, 100, 100, 100])

m = 2.0
c = 50.0
alpha = 0.005
epochs = 2000

n=len(x)
loss_history = []

m_history = []

for i in range(epochs):
    m_history.append(m)
    y_pred = m * x + c
    error = y - y_pred
    
    loss = np.mean(error ** 2)
    loss_history.append(loss)
    
    dm = -2 * np.mean(x * error)
    dc = -2 * np.mean(error)
    
    m -= alpha * dm
    c -= alpha * dc
    
print("기울기 m = ", m)
print("절편 c = ", c)
print(f"학습된 모델: y = {m}x + {c}")

x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + c

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x_line, y_line, color='red', label=f'Learned Line: y = {m:.2f}x + {c:.2f}')
plt.title('Linear Regression Trained with Gradient Descent')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(np.linspace(0, epochs, epochs), loss_history, color='green', label='MSE reduction curve')
plt.title('MSE reduction curve with number of iterations')
plt.xlabel('Iteratons')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.show()

print(f"학습 시간이 6.5시간일 때의 예상 성적 : {(m * 6.5 + c):.0f}점")
```
