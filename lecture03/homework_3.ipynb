다항 회귀(삼차)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 0.5, 1, 2, 3, 4, 4.5, 5, 5.5, 6, 7, 8, 9, 9.5, 10])
y = np.array([60, 62, 63, 68, 76, 83, 87, 89, 92, 94, 97, 99, 100, 100, 100])

m1 = 2.0
m2 = 2.0
m3 = 2.0
c = 50.0
alpha = 0.000001
epochs = 10000000

n=len(x)

for i in range(epochs):
    y_pred = m3 * (x ** 3) + m2 * (x ** 2) + m1 * x + c
    error = y - y_pred
    
    loss = np.mean(error ** 2)
    
    dm3 = -2 * np.mean((x ** 3) * error)
    dm2 = -2 * np.mean((x ** 2) * error)
    dm1 = -2 * np.mean(x * error)
    dc = -2 * np.mean(error)
    
    m1 -= alpha * dm1
    m2 -= alpha * dm2
    m3 -= alpha * dm3
    c -= alpha * dc
    
print("삼차항 계수 m3 = ", m3)
print("이차항 계수 m2 = ", m2)
print("일차항 계수 m1 = ", m1)
print("절편 c = ", c)
print(f"학습된 모델: y = {m3}x^3 + {m2}x^2 + {m1}x + {c}")

x_line = np.linspace(min(x), max(x), 100)
y_line = m3 * (x_line ** 3) + m2 * (x_line ** 2) + m1 * x_line + c

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x_line, y_line, color='red', label=f'Learned Line: y = {m3:.2f}x^3 + {m2:.2f}x^2 + {m1:.2f}x + {c:.2f}')
plt.title('Linear Regression Trained with Gradient Descent')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()

print(f"학습 시간이 6.5시간일 때의 예상 성적 : {(m3 * (6.5 ** 3) + m2 * (6.5 ** 2) + m1 * 6.5 + c):.0f}점")
```
